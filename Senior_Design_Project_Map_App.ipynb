```python
import pandas as pd
import folium

print("Adım 1: Veriler okunuyor ve akıllı sıralama yapılıyor")
df = pd.read_csv('Antalya_BigData_2019_2023.csv')

df = df.sort_values(by='YANGIN_DURUMU')

print("Adım 2: Kamera kilitli profesyonel harita altyapısı kuruluyor")
antalya_haritasi = folium.Map(
    location=[36.9081, 30.6956],
    zoom_start=9,
    min_zoom=8,
    max_bounds=True,
    min_lat=35.5, max_lat=37.8,
    min_lon=29.0, max_lon=33.0,
    tiles='OpenStreetMap'
)

fg_orman = folium.FeatureGroup(name="Yangın Çıkmamış Alanlar (2019-2023)")
fg_yangin = folium.FeatureGroup(name="Yangın Çıkan Bölgeler (2019-2023)")

print("Adım 3: Noktalar haritaya işleniyor...")
for index, row in df.iterrows():

    if row['YANGIN_DURUMU'] == 1:
        renk = '#e74c3c'
        durum_baslik = "GEÇMİŞ YANGIN BÖLGESİ" # Risk kelimesi kaldırıldı
        hedef_katman = fg_yangin
        cizgi_kalinligi = 2
        seffaflik = 0.9
    else:
        renk = '#2ecc71'
        durum_baslik = "YANGIN ÇIKMAMIŞ BÖLGE"
        hedef_katman = fg_orman
        cizgi_kalinligi = 0.5
        seffaflik = 0.3

    popup_icerik = f"""
    <div style="width:200px; font-family:Arial, sans-serif;">
        <h4 style="color:{renk}; margin-top:0; border-bottom:1px solid #ccc; padding-bottom:5px;">{durum_baslik}</h4>
        <b>Yıl:</b> {int(row['Yil'])}<br>
        <b>Yükseklik:</b> {row['Yukseklik']} m<br>
        <b>Eğim:</b> {row['Egim']:.1f}°<br>
        <b>Sıcaklık:</b> {row['Sicaklik_C']:.1f} °C<br>
        <b>Rüzgar:</b> {row['Ruzgar_Hizi']:.1f} m/s<br>
        <b>NDVI:</b> {row['NDVI']:.3f}
    </div>
    """

    folium.Circle(
        location=[row['Enlem'], row['Boylam']],
        radius=564,
        color=renk,
        weight=cizgi_kalinligi,
        fill=True,
        fill_color=renk,
        fill_opacity=seffaflik,
        popup=folium.Popup(popup_icerik, max_width=300)
    ).add_to(hedef_katman)

antalya_haritasi.add_child(fg_orman)
antalya_haritasi.add_child(fg_yangin)

folium.LayerControl(position='topright', collapsed=False).add_to(antalya_haritasi)

#başlık
baslik_html = '''
     <h3 align="center" style="font-size:20px; font-weight:bold; font-family:Arial, sans-serif;
     position:absolute; top:10px; left:50%; transform:translateX(-50%); z-index:9999;
     background-color:rgba(255,255,255,0.8); padding:10px; border-radius:5px; border:2px solid black;">
     Antalya Orman Yangınları Geçmiş Veri Haritası (2019-2023)
     </h3>
     '''
antalya_haritasi.get_root().html.add_child(folium.Element(baslik_html))

dosya_adi = 'Antalya_Gecmis_Veri_Haritasi.html'
antalya_haritasi.save(dosya_adi)

```
